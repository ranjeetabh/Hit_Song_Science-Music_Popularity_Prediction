```python
input_file = pd.read_csv('./Final_Files/sample_refined.csv', index_col=False)

# Shuffle data frame

input_file = input_file.sample(frac=1)

input_file = input_file.reset_index() 

input_file=input_file.drop(['index'],axis=1)

input_file.head(10)
```

```python
fileHeader = ["acousticness","analysis_url","danceability","duration_ms","energy","track_id",
              "instrumentalness","key","liveness","loudness","mode","speechiness","tempo","time_signature",
              "track_href","type","uri","valence","artist_name","track_name","popularity","bbhot", "artist_popularity","artist_followers"]

oFile = open("./Final_Files/sample_refined_updated.csv","w")

writer = csv.writer(oFile)
writer.writerow(fileHeader) 

```

```python
# Creating empty lists where the results are going to be stored

artist_followers = []
artist_count = 0

# Extracting artist popularity score for every track

for index, row in input_file.iterrows():
    
    row.artist_name = str(row.artist_name) 
    print("Artist_Name: ",row.artist_name)
    
    artist_results = sp.search(q=row.artist_name, type='artist')      
        
    if (artist_results['artists']['items']):    
        #print("here")
        for item in artist_results['artists']['items']:
            
            if item['followers']:
                row['artist_followers'] = item['followers']           
                pop = item['followers']    
                artist_count = artist_count + 1

                print("Artist_No: ",artist_count)
                print("\n")

                if pop:
                    break
                else:
                    item['followers'] = 0
            else:
                row['artist_followers'] = 0
    else:
        #print("No items")
        row['artist_followers'] = 0
            
    rowFields = [row['acousticness'],row['analysis_url'],row['danceability'],
                row['duration_ms'],row['energy'],row['track_id'],row['instrumentalness'],
                row['key'],row['liveness'],row['loudness'],row['mode'],row['speechiness'],
                row['tempo'],row['time_signature'],row['track_href'],row['type'],
                row['uri'],row['valence'],row['artist_name'],row['track_name'],row['popularity'],
                row['bbhot'],row['artist_popularity'],row['artist_followers']]
    
    writer.writerow(rowFields) 
 
oFile.close()
```
